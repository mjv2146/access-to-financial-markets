```python
# left merge on ind 
df = ind.merge(hh, on='hhid', how='inner')

# use level household weights as the main sample weight
df['weight'] = df['weight_hh']
```

```python
# create income_deciles
df['income_decile'] = pd.Categorical(utils.get_deciles(df['income'], weights=df['weight']))
df['income_quartile'] = utils.get_quartiles(df['income'], weights=df['weight'])
```

```python
to_tableau = ['age', 'sex', 'income', 'income_decile', 'education',
               'state', 'included']
special = {
    'naics':'Industry'
}

tableau_rename = {key: key.replace('_', ' ').title() for key in to_tableau}
tableau_rename.update(special)
tableau_rename
```

```python
os.chdir(processed)
df.to_csv('data.csv', index=False)
```

```python
df_tableau = df.copy()
df_tableau
```

```python
df_tableau['included'] = df_tableau['included'].astype(int)
```
