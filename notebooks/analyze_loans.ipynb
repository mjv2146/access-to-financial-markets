```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import os

dir = '/home/mitch/Dropbox/data/mexico_fls/interim/2005/'
os.chdir(dir)

data = pd.read_stata('loans_2005.dta')
```

```python
loan_types = data['where_get_loan'].unique()
data
```

```python
data.interest_rate_type.unique()
```

```python
data['charges_interest'] = data['interest_rate_type'].apply(lambda x : x in ['daily', 'monthly', 'annual'])
```

```python
import statsmodels.stats.weightstats as ws
charges_interest_by_type = {}
for loan_type in loan_types:
    focus = data.copy()[data['where_get_loan'] == loan_type]
    charges_interest_by_type[loan_type] = np.round(ws.DescrStatsW(focus['charges_interest'] + 0.0).mean, 2)
```
